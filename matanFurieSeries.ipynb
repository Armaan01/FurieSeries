```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_furie_elem(x,k,n):
    '''
    gets furie series value at point x with max num k
    n = 1 is for y=x
    n = 2 is for y=x^2
    '''
    sum = 0
    if n == 1:
        for i in range(1,k+1):
            sum+=-2/i*(-1)**i*np.sin(i*x)
    else:
        sum = (np.pi**2)/3
        for i in range(1,k+1):
            sum+=4*((-1)**(i))/(i**2)*np.cos(i*x)
    return sum

def main(num, pow):
    #num - number of furie seria
    x_dest = [-10 + k * 0.01 for k in range(2000)]
    y1_dest = [0]*len(x_dest)
    y2_dest = [0]*len(x_dest)
    for i in range(len(x_dest)):
        y1_dest[i] = get_furie_elem(x_dest[i],num,1)
        y2_dest[i] = get_furie_elem(x_dest[i],num,2)
    if pow == 1:
        plt.plot(x_dest, y1_dest)
    else:
        plt.plot(x_dest, y2_dest)

```

```python
main(1,1)
```
